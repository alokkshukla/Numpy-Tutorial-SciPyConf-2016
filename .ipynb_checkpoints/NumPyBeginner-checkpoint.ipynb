#### Multiple Figures

```python
t = linspace(0,2*pi,50)
x = sin(t)
y = cos(t)
```

```python
# Now create figure
figure()
plot(x)
show()
```

```python
figure()
plot(y)
show()
```

```python
# Dont click on figures, it makes them defualt
```

```python
figure(1)
plot(y)
show()
```

#### Multiple plots on same figure

#### Titles and Grids

```python
figure()
plot(x)
xlabel('radians')
ylabel('amplitude',fontsize='large')
title('Sin(x)')
grid();
```

```python
# Clear and close
plot(x)
clf()
# Close current
close()
close('all')
```

#### Error Bars 

```python
x = linspace(0,2*pi,10)
y = sin(x)
xerr = rand(10)
yerr = rand(10)/3
errorbar(x,y,yerr,xerr)
```
